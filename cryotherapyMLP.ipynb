```
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
seed = 7
np.random.seed(seed)
```

```
dataset = np.loadtxt("/content/sample_data/cryotherapy_dataset.csv",delimiter=",",skiprows=1)
X = dataset[:,0:6]
Y = dataset[:,6]
model = Sequential()
model.add(Dense(3,input_dim=6, init='uniform',activation='relu'))
model.add(Dense(2,init='uniform',activation='tanh'))
model.add(Dense(1,init='uniform',activation='sigmoid'))
history=model.compile(loss='mse',optimizer='RMSprop',metrics=['accuracy'])
history=model.fit(X,Y,nb_epoch=250,batch_size=100,validation_split=0.33)
print(history.history.keys())
```

```
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc = 'upper left')
plt.show()
```

```
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc = 'upper left')
plt.show()
scores = model.evaluate(X,Y)
print("\nAccuracy: %.2f%%" % (scores[1]*100))
```
